```python
import numpy as np
```

```python
while True:
  enter_vector = input("Please input product A with 5 properties(comma-seperated value):").split(",")
  A = np.array(enter_vector, dtype = float)
  if (len(A)!=5) or (A[3]<1) or (A[3]>10): 
    print("Error input!\n")
  else: break
print("Vector of A is: ",A)

while True:
  enter_vector = input("Please input product B with 5 properties(comma-seperated value):").split(",")
  B = np.array(enter_vector, dtype = float)
  if (len(B)!=5) or (B[3]<1) or (B[3]>10): 
    print("Error input!\n")
  else: break
print("Vector of A is: ",B)
```

```python
from numpy import dot
from numpy.linalg import norm
dot = dot(A,B)
cos_sim = dot/(norm(A)*norm(B))
degree = np.degrees(np.arccos(cos_sim))
print(f"Dot Product is : {round(dot,2)} and Angle two vectors are : {round(degree,2)} degree")
```

```python
import random
def GenerateProducts(num):
  return np.random.uniform(10.5, 75.5,size =(num,5))
data = GenerateProducts(200)
for i in range(200):
    data[i][3] = random.randint(1,10)
    for j in range(5):
      if j!=3:
        data[i][j] = round(data[i][j],2)
print("List 200 product with 5 properties are:")
print(data)
```

```python
def PCA_DimensionReducedef(X , num_components):
    #Step1
    X_meaned = X - np.mean(X , axis = 0)
    print(f"Mean = {X_meaned}")
    #Step2
    cov_mat = np.cov(X_meaned , rowvar = False)
    print(f"Cov = {cov_mat}")
    #Step3
    eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
    print(f"Eigen values = {eigen_values}")
    print(f"Eigne vectors = {eigen_vectors}")
    #Step4
    sorted_index = np.argsort(eigen_values)[::-1]
    sorted_eigenvalue = eigen_values[sorted_index]
    sorted_eigenvectors = eigen_vectors[:,sorted_index]
    #Step5
    eigenvector_subset = sorted_eigenvectors[:,0:num_components]
    #Step6
    X_reduced = np.dot(eigenvector_subset.transpose() , X_meaned.transpose() ).transpose()
    return X_reduced
result = PCA_DimensionReducedef(data,4)
print("List of random 200 products with 4 properties are",result)
print()

result = PCA_DimensionReducedef(data,3)
print("List of random 200 products with 3 properties are",result)
print()

result = PCA_DimensionReducedef(data,2)
print("List of random 200 products with 2 properties are",result)
print()

result = PCA_DimensionReducedef(data,1)
print("List of random 200 products with 1 properties are",result)
print()
```
